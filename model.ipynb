SPLITTING THE DATASET

Splitting into Matrix(X) and Target Vector(Y)

```python
#declaring variables
x = app[['Open']]
y = app[['High']]
```

```python
#importing train_test_split
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

USING LINEAR REGRESSION

```python
#importing LR
from sklearn.linear_model import LinearRegression

#Training the model
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#predictions 
y_pred= lr.predict(x_test)
y_pred
```

```python
#calculating mse 
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(x_test,y_pred)
mse
```

```python
#calculating mae
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(x_test,y_pred)
mae
```

```python
#plotting  predicted values using plotly
fig =px.line(y_pred)
fig.show()
```

```python
#plotting line plot
sns.distplot(y_pred)
plt.show()
```

USING LOGISTIC REGRESSION CLASSIFIER

```python
x=app[['Open']]
y=app[['Volume']]
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3)
```

```python
#importing LR
from sklearn.linear_model import LogisticRegression

#Training the model 
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
#calculating predictions 
predictions = classifier.predict(x_test)
predictions 
```

```python
#calculating mse
mse =mean_squared_error(x_test,predictions)
mse
```

```python
#calculating mae 
mae = mean_absolute_error (x_test,predictions)
mae
```

```python
#plotting line graph
fig=px.histogram (predictions)
fig.show()
```

We are done with the regression and classification techniques and we found some very accurate values of the predictions by the algorithms

USING RANDOM FOREST CLASSIFIER

```python
#importing RFC
from sklearn.ensemble import RandomForestClassifier

#Training the model
clf = RandomForestClassifier(n_estimators=100,random_state=42)
clf.fit(x_train,y_train)
```

```python
#predicting values
pred =clf.predict(x_test)
pred
```

```python
#Calculating mse 
mse =mean_squared_error(x_test,pred)
mse
```

```python
#Calculating mae
mae = mean_absolute_error(x_test,pred)
mae
```

```python
#Calculating rmse 
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error (x_test,pred)
rmse =np.sqrt(mse)
rmse
```

```python
#plotting the predicted values
sns.distplot(pred)
plt.show()
```

```python
fig =px.bar(pred)
fig.show()
```

USING DECISION TREE CLASSIFIER

```python
#importing DTC
from sklearn.tree import DecisionTreeClassifier

#Training the Model
DTC=DecisionTreeClassifier(random_state=42)
DTC.fit(x_train,y_train)
```

```python
#predicting 
predict=DTC.predict(x_test)
predict
```

```python
#Calculating mse 
mse = mean_squared_error(x_test,predict)
mse
```
